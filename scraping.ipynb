```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def fetch_html(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to fetch {url}: Status code {response.status_code}")
        return None

def extract_text_from_html(html):
    soup = BeautifulSoup(html, 'html.parser')
    return soup.get_text(separator='\n', strip=True)

def scrape_filtered_navigation_content(main_url):
    # Step 1: Fetch main page
    main_html = fetch_html(main_url)
    if not main_html:
        return

    # Step 2: Parse the main page to find relevant navigation links
    soup = BeautifulSoup(main_html, 'html.parser')

    # Define the keywords to filter links
    keywords = ['Publish', 'Home', 'Research', 'Biography', 'publications']

    # Step 3: Collect relevant links
    navigation_links = []
    nav_items = soup.find_all('a')  # Adjust this if your nav items are in a specific container
    for item in nav_items:
        link_text = item.get_text(strip=True)
        href = item.get('href')
        if href and any(keyword.lower() in link_text.lower() for keyword in keywords):  # Check if link text matches keywords
            full_url = urljoin(main_url, href)  # Handle relative URLs
            navigation_links.append(full_url)

    # Step 4: If no navigation links found, extract text from the main page
    if not navigation_links:
        print("No relevant navigation links found. Extracting text from the main page...")
        content = extract_text_from_html(main_html)
        print(f"\nContent from the main page:\n{content}")
    else:
        # Step 5: Fetch and extract text from each filtered linked page
        for link in navigation_links:
            linked_html = fetch_html(link)
            if linked_html:
                content = extract_text_from_html(linked_html)
                print(f"\nContent from {link}:\n{content}")
            else:
                print(f"Failed to retrieve content from {link}.")

# Example usage
main_url = input("Enter the main URL to scrape: ")
scrape_filtered_navigation_content(main_url)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def fetch_html(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to fetch {url}: Status code {response.status_code}")
        return None

def extract_text_from_html(html):
    soup = BeautifulSoup(html, 'html.parser')

    # Create a structured text output
    structured_text = ''

    # Process the HTML in the order of appearance
    for element in soup.body.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p', 'ul', 'ol','dl'], recursive=True):
        if element.name.startswith('h'):
            # Add heading
            structured_text += f"{element.name.upper()}: {element.get_text(strip=True)}\n"
        elif element.name == 'p':
            # Only add paragraphs that are not in an <li>
            if not element.find_parent(['li']):
                for content in element.contents:
                  if content.name == 'br':
                    structured_text += '\n'  # Add a line break for <br> tags
                  elif content.name != 'a':
                    # Add the text with a space before it if it's not the first element
                    if structured_text and structured_text[-1] not in ['\n', ' ']:
                        structured_text += ' '
                    structured_text += content.get_text(strip=True)  # Add text
                structured_text += '\n'  # Add an additional newline after the paragraph
        elif element.name in ['ul', 'ol']:
            # Add list items
            for item in element.find_all('li'):
                # Get text from li, excluding any p tags within li
                item_text = item.get_text(strip=True)
                structured_text += f"• {item_text}\n"
        elif element.name == 'dl':
            # Extract content from <dl> elements
            dl_text = element.get_text(strip=True)
            structured_text += f"• {dl_text}\n"

    return structured_text

def scrape_filtered_navigation_content(main_url):
    # Step 1: Fetch main page
    main_html = fetch_html(main_url)
    if not main_html:
        return

    # Step 2: Parse the main page to find relevant navigation links
    soup = BeautifulSoup(main_html, 'html.parser')

    # Define the keywords to filter links
    keywords = ['Publish', 'Home', 'Research', 'Biography', 'publications','projects']

    # Step 3: Collect relevant links
    navigation_links = []
    nav_items = soup.find_all('a')  # Adjust this if your nav items are in a specific container
    for item in nav_items:
        link_text = item.get_text(strip=True)
        href = item.get('href')
        if href and any(keyword.lower() in link_text.lower() for keyword in keywords):  # Check if link text matches keywords
            if urlparse(href).scheme and urlparse(href).netloc :
              break
            else:
              full_url = urljoin(main_url, href)  # Handle relative URLs, here if is a full then do not join
              navigation_links.append(full_url)

    # Step 4: If no navigation links found, extract text from the main page
    print("No relevant navigation links found. Extracting text from the main page...")
    outputfile=extract_text_from_html(main_html)

    content = extract_text_from_html(main_html)
    print(f"\nContent from the main page:\n{content}")

        # Step 5: Fetch and extract text from each filtered linked page
    used_link=[main_html]

    for link in navigation_links:
        linked_html = fetch_html(link)
        if linked_html and  linked_html not in used_link:
            content = extract_text_from_html(linked_html)
            outputfile = f"{outputfile}\n\n{'-' * 14}\n\n{content}"
            print(f"\nContent from {link}:\n{content}")
        else:
            print(f"Failed to retrieve content from {link}.")
    return outputfile


# Example usage
main_url = input("Enter the main URL to scrape: ")
out=scrape_filtered_navigation_content(main_url)
with open('f14.txt', 'w') as file:
    file.write(out)
```

加入了段落分割、空白检查和一些特殊的类别


```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def fetch_html(url):
    response = requests.get(url)
    # response = requests.get(url, verify=False)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to fetch {url}: Status code {response.status_code}")
        return None

def extract_text_from_html(html, separator='\n\n'):  #新增分割符号
    soup = BeautifulSoup(html, 'html.parser')


    # Create a structured text output
    structured_text = ''

    # Process the HTML in the order of appearance
    for element in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p', 'ul', 'ol', 'dl', 'td'], recursive=True):   #修改：去掉了.body；因为有的页面没有<body>
        if element.name.startswith('h'):
            # Add heading
            heading_text = f"{element.name.upper()}: {element.get_text(strip=True)}"
            if heading_text.strip():  # 新增空白检查
                structured_text += f"{heading_text}\n"
        elif element.name == 'p':
            # Only add paragraphs that are not in an <li>
            if not element.find_parent(['li']):
                has_content = False  # 新增变量，用于检查段落是否有非空内容
                for content in element.contents:
                    if content.name == 'br':
                        structured_text += '\n'  # Add a line break for <br> tags
                    elif content.name != 'a':
                        # Add the text with a space before it if it's not the first element
                        if structured_text and structured_text[-1] not in ['\n', ' ']:
                            structured_text += ' '
                        text = content.get_text(strip=True)
                        if text:  # 新增空白检查
                            has_content = True
                            structured_text += text  # Add text
                if has_content:  # 确保段落中至少有有效内容才添加分隔符
                    structured_text += f'\n{separator}\n'
        elif element.name in ['ul', 'ol']:
            # Add list items
            has_items = False  # 新增变量，用于检查列表是否有有效内容
            for item in element.find_all('li'):
                # Get text from li, excluding any p tags within li
                item_text = item.get_text(strip=True)
                if item_text:  # 新增空白检查
                    has_items = True
                    structured_text += f"• {item_text}\n"
            if has_items:  # 确保列表中有有效内容才添加分隔符
                structured_text += f"{separator}\n"
        elif element.name == 'dl':
            # Extract content from <dl> elements
            dl_text = element.get_text(strip=True)
            if dl_text.strip():  # 新增空白检查
                structured_text += f"• {dl_text}\n{separator}\n"
        elif element.name == 'td' and 'sectionHead' in element.get('class', []):
            table_text = element.get_text(strip=True)
            if table_text.strip():  # 新增空白检查
                structured_text += f"{table_text}\n"

    return structured_text.strip(f'\n{separator}\n')  # Remove the trailing separator

def scrape_filtered_navigation_content(main_url):
    # Step 1: Fetch main page
    main_html = fetch_html(main_url)
    if not main_html:
        return

    # Step 2: Parse the main page to find relevant navigation links
    soup = BeautifulSoup(main_html, 'html.parser')

    # Define the keywords to filter links
    keywords = ['Publish', 'Home', 'Research', 'Biography', 'publications','projects']

    # Step 3: Collect relevant links
    navigation_links = []
    nav_items = soup.find_all('a')  # Adjust this if your nav items are in a specific container
    for item in nav_items:
        link_text = item.get_text(strip=True)
        href = item.get('href')
        if href and any(keyword.lower() in link_text.lower() for keyword in keywords):  # Check if link text matches keywords
            if urlparse(href).scheme and urlparse(href).netloc :
              break
            else:
              full_url = urljoin(main_url, href)  # Handle relative URLs, here if is a full then do not join
              navigation_links.append(full_url)

    # Step 4: If no navigation links found, extract text from the main page
    print("No relevant navigation links found. Extracting text from the main page...")
    outputfile=extract_text_from_html(main_html)

    content = extract_text_from_html(main_html)
    print(f"\nContent from the main page:\n{content}")

        # Step 5: Fetch and extract text from each filtered linked page
    used_link=[main_html]

    for link in navigation_links:
        linked_html = fetch_html(link)
        if linked_html and  linked_html not in used_link:
            content = extract_text_from_html(linked_html)
            outputfile = f"{outputfile}\n\n{'-' * 14}\n\n{content}"
            print(f"\nContent from {link}:\n{content}")
        else:
            print(f"Failed to retrieve content from {link}.")
    return outputfile


# Exam'ple usage
main_url = input("Enter the main URL to scrape: ")
out=scrape_filtered_navigation_content(main_url)
with open('filename.txt', 'w') as file:
    file.write(out)

# from google.colab import files    #下载txt
# files.download("filename.txt")
```
